```python
import pandas as pd
import re
```

```python
df = pd.read_csv(r'C:\Users\wilko\Desktop\Studia\Projekty swoje\Python\Machine learning\NLP-text-classification\data\descriptions\opisy.csv', header=0)
```

```python
df.head()
```

```python
df_unfurnitured = df[df['Opis'].str.contains(r"\bbez mebli\b", case=False, regex=True) | 
            df['Opis'].str.contains(r"\bnie jest umeblowane\b", case=False, regex=True) |
            df['Opis'].str.contains(r"\bbrak mebli\b", case=False, regex=True) |
            df['Opis'].str.contains(r"\bnieumeblowane\b", case=False, regex=True) |
            df['Opis'].str.contains(r"\bbrak wyposażenia\b", case=False, regex=True) |
            df['Opis'].str.contains(r"\bbez umeblowania\b", case=False, regex=True) |
            df['Opis'].str.contains(r"\bnieumeblowana przestrzeń\b", case=False, regex=True) |
            df['Opis'].str.contains(r"\bnieurządzone\b", case=False, regex=True) |
            df['Opis'].str.contains(r"\bnie umeblowane\b", case=False, regex=True) | 
            df['Opis'].str.contains(r"\bbrak umeblowania\b", case=False, regex=True) |
            df['Opis'].str.contains(r"\bpuste\b", case=False, regex=True) |
            df['Opis'].str.contains(r"\bnie jest urządzone\b", case=False, regex=True) 
            ]
df_unfurnitured
```

```python
df_semifurnitured= df[df['Opis'].str.contains(r"\bczęściowo umeblowane\b", case=False, regex=True)]
df_semifurnitured
```

```python
pattern = r'\b(?!nieumeblowane|częściowo)\S*\s*umeblowane\b'
df_furnitured = df[df['Opis'].str.contains(pattern, case=False, regex=True)][:250]
df_furnitured
```

#### Opis klas
- nieumeblowane: 0
- umeblowane: 1
- częściowo umeblowane: 2

```python
df_unfurnitured['Label'] = df_unfurnitured['Label'].where(df_unfurnitured['Label'] == 'NaN', 0).astype(int) 
df_furnitured['Label'] = df_furnitured['Label'].where(df_furnitured['Label'] == 'NaN', 1).astype(int)
df_semifurnitured['Label'] = df_semifurnitured['Label'].where(df_semifurnitured['Label'] == 'NaN', 2).astype(int)
```

```python
df_combined = pd.concat([df_semifurnitured, df_furnitured, df_unfurnitured], ignore_index=True)
df_combined
```

```python
df_combined.to_csv(r'C:\Users\wilko\Desktop\Studia\Projekty swoje\Python\Machine learning\NLP-text-classification\data\descriptions\training_data.csv', index=None, header=True)
```
